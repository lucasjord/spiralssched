```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
with open('foo.txt') as w:
    times = [s for s in w.readlines() if '-' in s]
```

```python
slew = np.array([s.split() for s in times[0::2]])
earl = np.array([s.split() for s in times[1::2]])
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(earl[1:,-5].astype(float),'orange')
ax[0].plot(earl[1:,-4].astype(float),'g')
ax[0].plot(earl[1:,-3].astype(float),'r')
ax[0].plot(earl[1:,-2].astype(float),'m')
ax[0].plot(earl[1:,-1].astype(float),'b')
ax[0].set_ylabel('Early (s)');

ax[1].plot(slew[1:,-5].astype(float),'orange')
ax[1].plot(slew[1:,-4].astype(float),'g')
ax[1].plot(slew[1:,-3].astype(float),'r')
ax[1].plot(slew[1:,-2].astype(float),'m')
ax[1].plot(slew[1:,-1].astype(float),'b');
ax[1].set_ylabel('Slew (s)')
```
